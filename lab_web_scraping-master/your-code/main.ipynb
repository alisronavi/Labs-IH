#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
#your code
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
````

````python
languages1 = soup.find_all('div', {'class': 'central-featured-lang'})

languages2 = [label.strong.text for label in languages1]
articles = [label.small.text.replace('\xa0', '.') for label in languages1]

languages_articles = list(zip(languages2, articles))
languages_articles
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
#your code
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
table = soup.find_all('table', attrs = {'class':'wikitable'})[0]
rows = table.find_all('td')
columns = table.find_all('th')

language_table = [label.text.replace('\n', '')[:-3] if '[' in label.text.replace('\n', '') else label.text.replace('\n', '') for label in rows]
columns = [label.text.replace('\n', '')[:-3] if '[' in label.text.replace('\n', '') else label.text.replace('\n', '') for label in columns]

languages_dict = {columns[i]:language_table[i::len(columns)] for i in range(len(columns))}

languages_df = pd.DataFrame(languages_dict)
languages_df
````
